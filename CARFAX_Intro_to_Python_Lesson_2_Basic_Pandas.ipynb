## Create a list of all values in numerics['field'] using .to_list()

```python
numerics_list = numerics['field'].to_list()

numerics_list
```

## Loop over rows and columns to populate the table
#### Helpful Hint: You can produce a loop within a loop when you want to loop through multiple lists

```python
numerics_list = numerics['field'].to_list()

for n in numerics_list:
    for f in fields:
        nxt = numerics['field'] == n
        if f == 'min':
            numerics.loc[nxt,f] = data1[n].min()
        if f == 'mean':
            numerics.loc[nxt,f] = data1[n].mean()
        if f == 'max':
            numerics.loc[nxt,f] = data1[n].max()

numerics
```

## Create a list comprehension to find the metrics we need
#### Helpful Hint: List comprehension is a good way to create lists
#### [expression for item in iterable if condition]

```python
facts = [i for i in data1.columns if 'ee_' in i or 'ep_' in i or 'incloss_' in i or 'cc_' in i]
facts
```

## Define agg_dict
## Create a function eda_table0 that loops through the facts list and produces a table of aggregated data.

```python
agg_dict = {i: 'sum' for i in facts}

def eda_table0(data,field):
    a = data.groupby([field], dropna = False).agg(agg_dict).reset_index()

    for f in facts:
        a[f + '_pct'] = a[f]/a[f].sum()

    return a
```

## Set Display Settings to determine the number of columns and rows viewed and the size of the display.

## Run eda_table0 function on data1 dataset for cbistier

```python
pd.set_option("display.max_columns",25)
pd.set_option("display.max_rows",250)
pd.options.display.float_format = '{:,.2f}'.format

eda_table0(data1,'cbistier')
```

## Create a more specific function, eda_table

```python
facts = [i for i in data1.columns if 'ee_' in i or 'ep_' in i or 'incloss_' in i or 'cc_' in i]
ee_list = [i for i in data1.columns if 'ee_' in i]
agg_dict = {i: 'sum' for i in facts}

covs = ['bi','col']

def eda_table(data,field):
    a = data.groupby([field], dropna = False).agg(agg_dict).reset_index()

    for e in ee_list:
        a[e + '_pct'] = a[e]/a[e].sum()

    for c in covs:
        a['avg_prem_' + c] = a['ep_' + c]/a['ee_' + c]
        a['lr_' + c] = a['incloss_' + c]/a['ep_' + c]
        a['freq_' + c] = a['cc_' + c]/a['ee_' + c]
        a['sev_' + c] = a['incloss_' + c]/a['cc_' + c]
        a['pp_' + c] = a['incloss_' + c]/a['ee_' + c]

    a.drop(columns = facts, inplace = True)
#     a = a.drop(columns = facts)

    return a
```

## Run eda_table function on data1 for cbistier

```python
eda_table(data1,'cbistier')
```

## Create a loop to run data1 through eda_table function for all j in list comprehension

#### Hint: Use print('------') at end of loop segment each iteration in the display

```python
for j in [i for i in data1.columns if i not in facts]:
    display(eda_table(data1,j))
    print('-------------------------------------------------------------------------------------------------------------------------------------------------')
```
